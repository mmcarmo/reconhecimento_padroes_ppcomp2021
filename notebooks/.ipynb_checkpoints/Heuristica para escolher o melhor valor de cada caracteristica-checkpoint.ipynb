## A classe MyTree foi criada para experimentos deste exercício, e é uma implementação recursiva de uma arvore de decisão simples.

```python
# CODE BASED ON https://scikit-learn.org/stable/developers/develop.html
# Codigo da arvore baseado nos exercicios propostos em video aula e no github do professor Boldt
# https://github.com/fboldt/aulasml/blob/master/arvore.ipynb

from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.utils.validation import check_X_y, check_array, check_is_fitted
from sklearn.utils.multiclass import unique_labels
from sklearn.metrics import euclidean_distances
from collections import Counter
from scipy import stats


class MyTree(BaseEstimator, ClassifierMixin):

    def __init__(self, use_boldt_heuristic=False, n_samples=7):
        self.use_boldt_heuristic = use_boldt_heuristic
        self.n_samples = n_samples

    def fit(self, X, y):

        # Check that X and y have correct shape
        X, y = check_X_y(X, y)
        # Store the classes seen during fit
        self.classes_ = unique_labels(y)

        self.X_ = X
        self.y_ = y
        # Return the classifier
        
        # SELETOR DE HEURISTICA DE ESCOLHA DA MELHOR CARACTERISTICA
        if self.use_boldt_heuristic:
            # GULOSO
            self.caracteristica, self.valor, self.imp = self.melhorCaracteristica(X,y)
        else:
            # RANDOM com n_samples
            self.caracteristica, self.valor, self.imp = self.melhorCaracteristicaRandom(X,y,self.n_samples)
        
        maiores = X[:,self.caracteristica] > self.valor
        if sum(maiores)>0 and sum(~maiores)>0:
            self.maiores = MyTree(use_boldt_heuristic=self.use_boldt_heuristic,n_samples=self.n_samples)
            self.maiores.fit(X[maiores,:],y[maiores])
            self.menores = MyTree(use_boldt_heuristic=self.use_boldt_heuristic,n_samples=self.n_samples)
            self.menores.fit(X[~maiores,:],y[~maiores])
        else:
            self.resposta = self.maisFrequente(y)
        
        return self

    def predict(self, X):

        # Check is fit had been called
        check_is_fitted(self)

        # Input validation
        #X = check_array(X)

        y = np.empty((X.shape[0]))
        if hasattr(self, 'resposta'):
            y[:] = self.resposta
        else:
            maiores = X[:,self.caracteristica] > self.valor
            y[maiores] = self.maiores.predict(X[maiores,:])
            y[~maiores] = self.menores.predict(X[~maiores,:])
        return y

    
    @staticmethod
    def maisFrequente(y):
        return Counter(y.flat).most_common(1)[0][0]
    
    @staticmethod
    def impureza(y):
        l = sorted(list(set(y)))
        prob = np.zeros(len(l))
        for i,k in enumerate(l):
            prob[i] = sum(y==k)/len(y)
        
        imp = 1 - sum(prob**2)
        return imp
    
    #@staticmethod
    def impurezaValor(self, x, y, valor, categorical=False):
        if categorical:
            iguais = x==valor
            impurezaIguais = self.impureza(y[iguais])
            propIguais = sum(iguais)/len(y)
            impurezaDiferentes = self.impureza(y[~iguais])
            propDiferentes = sum(~iguais)/len(y)
            impurezaTotal = propIguais*impurezaIguais + propDiferentes*impurezaDiferentes
            return impurezaTotal, impurezaIguais, impurezaDiferentes
        else:
            maiores = x>valor
            impurezaMaiores = self.impureza(y[maiores])
            propMaiores = sum(maiores)/len(y)
            impurezaMenores = self.impureza(y[~maiores])
            propMenores = sum(~maiores)/len(y)
            impurezaTotal = propMaiores*impurezaMaiores + propMenores*impurezaMenores
            return impurezaTotal, impurezaMaiores, impurezaMenores
        
    #@staticmethod
    def melhorValor(self, x, y):
        result = None
        menorImpureza = np.inf
        xmax = np.max(x)
        xmin = np.min(x)
        while True:
            valor = (xmin+xmax)/2
            impTotal, impMaiores, impMenores = self.impurezaValor(x, y, valor)
            if impTotal < menorImpureza:
                menorImpureza = impTotal
                result = valor
                if impMaiores == 0 or impMenores == 0:
                    break
                if impMaiores < impMenores:
                    xmin = valor
                else:
                    xmax = valor
            else:
                break
        return result, menorImpureza

    #@staticmethod
    def melhorCaracteristica(self,X, y):
        impurezas = []
        valores = []
        for caracteristica in range(X.shape[1]):
            valor, imp = self.melhorValor(X[:,caracteristica], y)
            impurezas.append(imp)
            valores.append(valor)
        impurezas = np.array(impurezas)
        caracteristica = np.argmin(impurezas)
        return caracteristica, valores[caracteristica], impurezas[caracteristica]
    
    def melhorValorRandom(self, x ,y,n_samples=7):
        result = None
        menorImpureza = np.inf
        xmax = np.max(x)
        xmin = np.min(x)
        samples = np.random.uniform(low=xmin,high=xmax,size=n_samples)
        for value in samples:
            impTotal, impMaiores, impMenores = self.impurezaValor(x, y, value)
            if impTotal < menorImpureza:
                menorImpureza = impTotal
                result = value
        
        return result, menorImpureza
    
    def melhorCaracteristicaRandom(self,X, y,n_samples=7):
        impurezas = []
        valores = []
        if n_samples > X.shape[1]: n_samples=X.shape[1]
        samples = np.random.choice(X.shape[1],n_samples,replace=False)
        #print(samples)
        for caracteristica in samples:
            valor, imp = self.melhorValorRandom(X[:,caracteristica], y,n_samples)
            impurezas.append(imp)
            valores.append(valor)
        impurezas = np.array(impurezas)
        min_impureza = np.argmin(impurezas)
        caracteristica = samples[min_impureza]
        return caracteristica, valores[min_impureza], impurezas[min_impureza]
```

## Resolvemos melhorar a função de visualização com uma simples redução dimencional para visualização em duas dimensões. Neste dataset um simples PCA consegue capturar toda a variabilidade em somente 2 dimensões perfeitamente.

```python
import matplotlib.pyplot as plt
from umap import UMAP
from sklearn.decomposition import PCA

######################################################################
# Pequena Modificação da função de impressão passada em aula online para visualizar espaços multidimencionais em
# um espaço de dimensões reduzidas para somente 2. O algoritmo padrão é o PCA,
# mas também pode ser optado pelo UMAP por parametro (os resultados do UMAP são menos intuitivos, mas podem ser
# mais explicativos que o PCA(2))
#####################################################################

def plotDecisao(modelo, X, y,use_umap=False):
    fig, ax = plt.subplots(figsize=(14, 8))
    modelo.fit(X, y)
    scores = cross_validate(modelo, X, y)
    print('CrossValidation Scores:')
    print(scores['test_score'], np.mean(scores['test_score']))
    if use_umap: emb = UMAP(n_components=2,random_state=42)
    else: emb = PCA(n_components=2,random_state=42)
    X_emb = emb.fit_transform(X)
    print("passei pelo treino e redução dimencional...")
    x0s = np.linspace(np.min(X_emb[:,0])-0.2, np.max(X_emb[:,0])+0.2, 100)
    x1s = np.linspace(np.min(X_emb[:,1])-0.2, np.max(X_emb[:,1])+0.2, 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    Xdec = np.c_[x0.ravel(), x1.ravel()]
    print("agora vem o inverse decode...")
    Xinv = emb.inverse_transform(Xdec) # Reconstrucão de dados do espaco reduzido
    #print(Xinv)
    ypred = modelo.predict(Xinv) # Predicao dos dados reconstruidos do espaco reduzido
    ax.contourf(x0, x1, ypred.reshape(x0.shape), alpha=0.25)
    for k in set(y):
        ax.plot(X_emb[:,0][y==k], X_emb[:,1][y==k], 'o', label=k)
    
    ax.legend()
    plt.xlabel(f"X_emb[0]")
    plt.ylabel(f"X_emb[1]")
    plt.show()

```

```python
%%time
# Arvore padrão, conforme sugestão de implementação nas video aulas
plotDecisao(MyTree(use_boldt_heuristic=True),X, y)
```

```python
%%time
#Arvore com heuristica de seleção de n_samples aleatórios com a menor impureza
plotDecisao(MyTree(use_boldt_heuristic=False,n_samples=3),X, y)
```

```python
# Comprarativo com o kNN padrão
plotDecisao(KNeighborsClassifier(),X, y)
```

```python
# Comparativo com a implementação padrão da arvore de desição do Scikit-Learn
plotDecisao(DecisionTreeClassifier(),X, y)
```

# Resultados do experimento 1:

### Neste experimento a decisão aleatória foi potencialmente tão boa quanto as demais implementações de arvores de decisão.
