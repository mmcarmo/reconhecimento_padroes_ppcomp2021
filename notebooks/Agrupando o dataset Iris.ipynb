```python
import numpy as np
import pandas as pd
import sklearn

import matplotlib.pyplot as plt

np.__version__,pd.__version__,sklearn.__version__
```

```python
from sklearn.datasets import load_iris
data = load_iris()
X = data.data[:,2:4]
y = data.target
data.target_names
```

```python
for i in range(len(set(y))):
    plt.plot(X[y==i, 0], X[y==i, 1], ".", label=data.target_names[i])
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.legend()
plt.show()
```

```python
plt.scatter(X[:,0], X[:,1], marker=".")
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
print(kmeans.cluster_centers_)

plt.scatter(X[:,0], X[:,1], marker=".")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="x", color="r", s=100, linewidths=3)
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (12,8)
for i in range(len(set(y))):
    plt.plot(X[y==i, 0], X[y==i, 1], "o", label=data.target_names[i])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="*", color="b", s=100, linewidths=3)
ypred = kmeans.predict(X)
for i in range(len(set(ypred))):
    if i==0:
        plt.plot(X[ypred==i, 0], X[ypred==i, 1], "x", label=data.target_names[i])
    elif i==1:
        plt.plot(X[ypred==i, 0], X[ypred==i, 1], "s", label=data.target_names[i])
    else:
        plt.plot(X[ypred==i, 0], X[ypred==i, 1], ".", label=data.target_names[i])
#    if ypred[i] != data.target_names[i]:
#        print(ypred[i],data.target_names[i])
#        plt.scatter(X[i,0],X[i,1],marker='x',color='r',s=50,linewidths=3)
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.legend()
plt.show()
```
